```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the Wine dataset
wine_data = pd.read_csv("/content/drive/MyDrive/dataset/wine_data.csv")

# Extract features from the dataset (assuming the Wine dataset structure and excluding the label column)
x = wine_data.iloc[:, :-1].values

# Initialize an empty list to store the within-cluster sum of squares (WCSS)
wcss = []

# Run K-means for different numbers of clusters (from 1 to 10) and compute WCSS
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

# Plot the WCSS values to find the optimal number of clusters using the Elbow Method
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load the Wine dataset
wine_data = pd.read_csv("/content/drive/MyDrive/dataset/wine_data.csv")

# Extract features from the dataset (assuming the Wine dataset structure and excluding the label column)
x = wine_data.iloc[:, :-1].values

# Apply PCA to reduce dimensions to 2 for visualization
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x)

# Apply K-means clustering with 3 clusters on the reduced data
kmeans_pca = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans_pca = kmeans_pca.fit_predict(x_pca)

# Plot the clusters and centroids
plt.scatter(x_pca[y_kmeans_pca == 0, 0], x_pca[y_kmeans_pca == 0, 1], s=20, c='purple', label='Cluster 1')
plt.scatter(x_pca[y_kmeans_pca == 1, 0], x_pca[y_kmeans_pca == 1, 1], s=20, c='orange', label='Cluster 2')
plt.scatter(x_pca[y_kmeans_pca == 2, 0], x_pca[y_kmeans_pca == 2, 1], s=20, c='green', label='Cluster 3')

# Plot the centroids of the clusters
plt.scatter(kmeans_pca.cluster_centers_[:, 0], kmeans_pca.cluster_centers_[:, 1], s=100, c='red', label='Centroids')

plt.title('Clusters of Wine Data (PCA Reduced)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```
