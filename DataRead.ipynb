```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the Wine dataset
wine_data = pd.read_csv("/content/drive/MyDrive/dataset/wine_data.csv")

# Extract features from the dataset (assuming the Wine dataset structure)
x = wine_data.iloc[:, :-1].values  # Typically exclude the last column if it is a label

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(x)

# Plot the clusters and centroids
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s=20, c='purple', label='Cluster 1')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s=20, c='orange', label='Cluster 2')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s=20, c='green', label='Cluster 3')

# Plot the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red', label='Centroids')

plt.title('Clusters of Wine Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
x_pca = pca.fit_transform(x)

# Apply K-means clustering on the reduced data
y_kmeans_pca = kmeans.fit_predict(x_pca)

# Plot the clusters and centroids
plt.scatter(x_pca[y_kmeans_pca == 0, 0], x_pca[y_kmeans_pca == 0, 1], s=20, c='purple', label='Cluster 1')
plt.scatter(x_pca[y_kmeans_pca == 1, 0], x_pca[y_kmeans_pca == 1, 1], s=20, c='orange', label='Cluster 2')
plt.scatter(x_pca[y_kmeans_pca == 2, 0], x_pca[y_kmeans_pca == 2, 1], s=20, c='green', label='Cluster 3')

# Plot the centroids of the clusters in the reduced space
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red', label='Centroids')

plt.title('Clusters of Wine Data (PCA Reduced)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```
